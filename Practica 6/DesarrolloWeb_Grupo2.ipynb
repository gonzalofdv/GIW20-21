### Práctica sobre APIs

Se va a explotar la API REST vista en clase sobre películas y series OMDb API que se encuentra en:

http://www.omdbapi.com/

Lo primero que se debe hacer es obtener una clave de uso en el siguiente lugar:

Se pide crear una aplicación que actúe como un buscador con varias opciones:
 
 * __Opción 1__: Buscar por nombre de película. El usuario podrá introducir el nombre de una película y se deberá mostrar de forma amigable lo recuperado por cada película reconocida.
 
 * __Opción 2__: Buscar por identificador de película. El usuario podrá introducir elidentificador de una película y se deberá mostrar de forma amigable lo recuperado por cada película reconocida.
 
 * __Opción 3__: Búsqueda ampliada. Es una combinación de las opciones 1 y 2.El usuario introduce el nombre de una película y por cada película recuperada se deberá mostrar la información recuperada más la información adicional que se puede recuperar de cada película usando el identificador de pelicula. 
 
 * __Opción 4__: Recuperacion de posters.El usuario podrá introducir el nombre de una película y se deberá mostrar por pantalla el nombre de cada película recuperada y la dirección al poster. Además se creará una carpeta donde se almacenarán los posters que deberán ser recuperados a partir de las direcciones indicadas.

 * __Opción 5__: Recuperacion con poster.El usuario podrá introducir el nombre de una película y se deberá mostrar por pantalla los nombres de las diferentes películas recuperadas. A continuación, el usuario elegirá una de esas películas y entonces se le mostrará por pantalla toda la información ampliada sobre esa película (como en la opción 2) pero además se le mostrará el poster de la película (para ello hay que bajarse previamente el poster de la película y usar el método Image como se realiza en los apuntes.
 * __Opción 6__: Salir o continuar.

```python
from PIL import Image
import os
import json
import requests
import urllib.request
from urllib.parse import urlencode, quote_plus
clave="8c3d1f10" #Clave de la api tras el registro
uri="http://www.omdbapi.com/?apikey="+clave #uri

   
    
#*************************  Opción 1  *******************************

def busquedaPeli():
    print("Introduce el título de la película que quieres buscar:")
    titulo = input()
    r = requests.get(uri)
    url = uri + "&" + urlencode({"s":titulo, "type":"movie"}, quote_via=quote_plus) #consulta
    manf = urllib.request.urlopen(url)
    resultados = []
    
    for x in manf.readlines():
        dic = json.loads(x) #cargamos el json en un diccionario
        
    if (dic["Response"] == "False"): #Si no hay resultados
        print("\nHa habido algún problema con tu búsqueda")
        print("Error: " + dic["Error"] + "\n") #Mostramos el error que nos ofrece la API
    else:
        print("Los resultados obtenidos son los siguientes:\n")

        for peli in dic["Search"]: #la respuesta al buscar por titulo es un json con clave search y value otro json con las peliculas
            print("---------------------------------------")
            print("Titulo: " + peli["Title"] + "")
            print("imdbID: " + peli["imdbID"] + "")
            print("Tipo: " + peli["Type"] + "")
            print("Poster: " + peli["Poster"] + "")
            print("---------------------------------------\n")

#*************************  Opción 2  *******************************


def busquedaId():
    print("Introduce el identificador de la película que quieres consultar:")
    iden = input()
    r = requests.get(uri)
    url = uri + "&"+urlencode({"i":iden},quote_via=quote_plus)
    manf = urllib.request.urlopen(url)
    
    string = manf.read().decode('utf-8')
    peli = json.loads(string) #Al buscar por ID nos devuelve directamente el json de la película
    
    if (peli["Response"] == "False"): #Si no hay resultados
        print("\nHa habido algún problema con tu búsqueda")
        print("Error: " + peli["Error"] + "\n") #Mostramos el error que nos ofrece la API
    else:
        print("---------------------------------------")
        print("Titulo: " + peli["Title"] + "\n")
        print("Anio: " + peli["Year"] + "\n")
        print("Restricción de edad: " + peli["Rated"] + "\n")
        print("Lanzamiento: " + peli["Released"] + "\n")
        print("Duración: " + peli["Runtime"] + "\n")
        print("Género: " + peli["Genre"] + "\n")
        print("Director/es: " + peli["Director"] + "\n")
        print("Escritor/es: " + peli["Writer"] + "\n")
        print("Reparto: " + peli["Actors"] + "\n")
        print("Argumento: " + peli["Plot"] + "\n")
        print("Idioma: " + peli["Language"] + "\n")
        print("País: " + peli["Country"] + "\n")
        print("Premios: " + peli["Awards"] + "\n")
        print("Poster: " + peli["Poster"] + "\n")
        print("Valoraciones: ")
        for valoracion in peli["Ratings"]: #Recorremos las distintas valoraciones
            print("Fuente: " + valoracion["Source"] + "  Valoracion: " + valoracion["Value"])
        print("\nMetascore: " + peli["Metascore"] + "\n")
        print("imdbRating: " + peli["imdbRating"] + "\n")
        print("imdbVotes: " + peli["imdbVotes"] + "\n")
        print("imdbID: " + peli["imdbID"] + "\n")
        print("Tipo: " + peli["Type"] + "\n")
        print("DVD: " + peli["DVD"] + "\n")
        print("BoxOffice: " + peli["BoxOffice"] + "\n")
        print("Productora: " + peli["Production"] + "\n")
        print("Sitio web: " + peli["Website"] + "\n")
        print("---------------------------------------\n")
    
    
#*************************  Opción 3  *******************************

def busquedaAmpliada():
    print("Introduce el título de la película que quieres buscar:")
    titulo = input()
    r = requests.get(uri)
    url = uri + "&" + urlencode({"s":titulo, "type":"movie"}, quote_via=quote_plus)
    manf = urllib.request.urlopen(url)
    resultados = []
    
    for x in manf.readlines():
        dic = json.loads(x)
    
    if (dic["Response"] == "False"): #Si no hay resultados
        print("\nHa habido algún problema con tu búsqueda")
        print("Error: " + dic["Error"] + "\n") #Mostramos el error que nos ofrece la API
    else:
    
        print("Los resultados obtenidos son los siguientes:\n")

        for peli in dic["Search"]:

            r2 = requests.get(uri)
            url2 = uri + "&"+urlencode({"i":peli["imdbID"]},quote_via=quote_plus) #A partir del ID que nos da la busqueda por titulo, realizamos la busqueda ampliada
            manf2 = urllib.request.urlopen(url2)

            string = manf2.read().decode('utf-8')
            peliA = json.loads(string)

            print("---------------------------------------")
            print("Titulo: " + peli["Title"] + "")
            print("imdbID: " + peli["imdbID"] + "")
            print("Anio: " + peliA["Year"] + "\n")
            print("Restricción de edad: " + peliA["Rated"] + "\n")
            print("Lanzamiento: " + peliA["Released"] + "\n")
            print("Duración: " + peliA["Runtime"] + "\n")
            print("Género: " + peliA["Genre"] + "\n")
            print("Director/es: " + peliA["Director"] + "\n")
            print("Escritor/es: " + peliA["Writer"] + "\n")
            print("Reparto: " + peliA["Actors"] + "\n")
            print("Argumento: " + peliA["Plot"] + "\n")
            print("Idioma: " + peliA["Language"] + "\n")
            print("País: " + peliA["Country"] + "\n")
            print("Premios: " + peliA["Awards"] + "\n")
            print("Poster: " + peliA["Poster"] + "\n")
            print("Valoraciones: ")
            for valoracion in peliA["Ratings"]:
                print("Fuente: " + valoracion["Source"] + "  Valoracion: " + valoracion["Value"])
            print("\nMetascore: " + peliA["Metascore"] + "\n")
            print("imdbRating: " + peliA["imdbRating"] + "\n")
            print("imdbVotes: " + peliA["imdbVotes"] + "\n")
            print("imdbID: " + peliA["imdbID"] + "\n")
            print("Tipo: " + peliA["Type"] + "\n")
            print("DVD: " + peliA["DVD"] + "\n")
            print("BoxOffice: " + peliA["BoxOffice"] + "\n")
            print("Productora: " + peliA["Production"] + "\n")
            print("Sitio web: " + peliA["Website"] + "\n")
            print("---------------------------------------\n")

#*************************  Opción 4  *******************************

def descargaPoster():
    print("Introduce el título de la película que quieres buscar:")
    titulo = input()
    r = requests.get(uri)
    url = uri + "&" + urlencode({"s":titulo, "type":"movie"}, quote_via=quote_plus)
    manf = urllib.request.urlopen(url)
    resultados = []
    
    for x in manf.readlines():
        dic = json.loads(x)
    
    if (dic["Response"] == "False"): #Si no hay resultados
        print("\nHa habido algún problema con tu búsqueda")
        print("Error: " + dic["Error"] + "\n") #Mostramos el error que nos ofrece la API
    else:
    
        print("Los resultados obtenidos son los siguientes:\n")

        #Creacion del directorio para guardar los posters
        os.makedirs("posters", exist_ok=True)

        for peli in dic["Search"]:
            print("---------------------------------------")
            print("Titulo: " + peli["Title"] + "")
            print("Poster: " + peli["Poster"] + "")
            print("---------------------------------------\n")
            if peli["Poster"] != "N/A": #Si hay poster, lo guardaremos en la carpeta correspondiente y le asignamos un nombre
                img = urllib.request.urlopen(peli["Poster"]).read()
                manf2 = open("posters/"+peli["imdbID"]+".jpg", "wb") #guardamos poster con el nombre del ID de pelicula
                manf2.write(img)
                manf2.close
            
#*************************  Opción 5  *******************************

def muestraPoster():
    print("Introduce el título de la película:")
    titulo = input()
    r = requests.get(uri)
    url = uri + "&" + urlencode({"s":titulo, "type":"movie"}, quote_via=quote_plus)
    manf = urllib.request.urlopen(url)
    resultados = []
    
    for x in manf.readlines():
        dic = json.loads(x)
    
    if (dic["Response"] == "False"): #Si no hay respuesta 
        print("\nHa habido algún problema con tu búsqueda")
        print("Error: " + dic["Error"] + "\n") #Mostramos el error que nos devuelve la API
    else:
    
        print("Elige una de las siguientes películas:\n")

        i = 1
        for peli in dic["Search"]: #Hacemos print de los titulos encontrados con un número asignado
            print(str(i) + ". Título: " + peli["Title"])
            i = i + 1

        eleccion = int(input())

        r2 = requests.get(uri)
        url2 = uri + "&"+urlencode({"i":dic["Search"][eleccion-1]["imdbID"]},quote_via=quote_plus) #Buscamos la info ampliada del titulo seleccionado
        manf2 = urllib.request.urlopen(url2)


        string = manf2.read().decode('utf-8')
        peliA = json.loads(string)

        #Si no existe el directorio, se crea
        os.makedirs("posters", exist_ok=True)

        if peliA["Poster"] != "N/A": #Si la película dispone de poster
                img = urllib.request.urlopen(peliA["Poster"]).read()
                manf3 = open("posters/"+peliA["imdbID"]+".jpg", "wb") #guardamos poster con el nombre del ID de pelicula
                manf3.write(img)
                manf3.close
                imagen = Image.open("posters/"+peliA["imdbID"]+".jpg")
                imagen.show() #Mostramos la imagen con el visor por defecto de Windows, siempre que la pelicula tenga asignada un poster
    
        
        print("---------------------------------------")
        print("Titulo: " + peliA["Title"] + "")
        print("imdbID: " + peliA["imdbID"] + "")
        print("Anio: " + peliA["Year"] + "\n")
        print("Restricción de edad: " + peliA["Rated"] + "\n")
        print("Lanzamiento: " + peliA["Released"] + "\n")
        print("Duración: " + peliA["Runtime"] + "\n")
        print("Género: " + peliA["Genre"] + "\n")
        print("Director/es: " + peliA["Director"] + "\n")
        print("Escritor/es: " + peliA["Writer"] + "\n")
        print("Reparto: " + peliA["Actors"] + "\n")
        print("Argumento: " + peliA["Plot"] + "\n")
        print("Idioma: " + peliA["Language"] + "\n")
        print("País: " + peliA["Country"] + "\n")
        print("Premios: " + peliA["Awards"] + "\n")
        print("Poster: " + peliA["Poster"] + "\n")
        print("Valoraciones: ")
        for valoracion in peliA["Ratings"]:
            print("Fuente: " + valoracion["Source"] + "  Valoracion: " + valoracion["Value"])
        print("\nMetascore: " + peliA["Metascore"] + "\n")
        print("imdbRating: " + peliA["imdbRating"] + "\n")
        print("imdbVotes: " + peliA["imdbVotes"] + "\n")
        print("imdbID: " + peliA["imdbID"] + "\n")
        print("Tipo: " + peliA["Type"] + "\n")
        print("DVD: " + peliA["DVD"] + "\n")
        print("BoxOffice: " + peliA["BoxOffice"] + "\n")
        print("Productora: " + peliA["Production"] + "\n")
        print("Sitio web: " + peliA["Website"] + "\n")
        print("---------------------------------------\n")

        


#*************************  Menu principal  ************************* 
opc = 10 #Valor auxiliar para iniciar el programa
while (opc != 6): 
    print("Elige una de las siguientes opciones: ")
    print("1.- Buscar por nombre de película")
    print("2.- Buscar por identificador de película")
    print("3.- Búsqueda ampliada")
    print("4.- Recuperación de posters")
    print("5.- Recuperación con póster")
    print("6.- Salir")
    opc = int(input())
    if opc == 1:
        busquedaPeli()
    elif opc == 2:
        busquedaId()
    elif opc == 3:
        busquedaAmpliada()
    elif opc == 4:
        descargaPoster()
    elif opc == 5:
        muestraPoster()
```
